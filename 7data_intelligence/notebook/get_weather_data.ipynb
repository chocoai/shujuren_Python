```python
import requests
```

```python
import random
```

```python
url = "http://www.weather.com.cn/weather/101010100.shtml"
```

```python
timeout = random.choice(range(80, 180))
print(timeout)
```

```python
req_content = requests.get(url, timeout=timeout)
```

```python
req_content.encoding = 'utf-8'
```

```python
status_code = req_content.status_code
```

```python
print(status_code)
```

```python
print(req_content.text)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(req_content.text, "html.parser")
```

```python
ul_tag = soup.find("ul", "t clearfix")
```

```python
print(ul_tag)
```

```python
li_tag = ul_tag.findAll("li")
```

```python
for tag in li_tag:
    print(tag.find("h1").string)
    print(tag.find("p", "wea").string)
    try:
        print(tag.find("p", "tem").find("span").string)
        print(tag.find("p", "tem").find("i").string)
    except:
        print("没有高温或者低温数据")
        pass
    print(tag.find("p", "win").find("i").string)
    print("_______________ 分割线 ____________________")
```
