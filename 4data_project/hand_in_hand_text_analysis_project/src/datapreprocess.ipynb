```python
import os
import pandas as pd
from snownlp import SnowNLP as snlp
import re
```

```python
data_path="../data/raw/"
```

```python
# 获取语料函数
def get_corpus(textname):
    content=''
    with open(data_path+textname, 'rb') as f:
        while True:
            try:
                temp=f.readline()
                if len(temp)==0:
                    break
                else:
                    content+=temp.decode("gb2312")
            except Exception:
                continue
    #利用正则表达式过滤不要的内容
    quote_pat=re.compile(r"，|。|！|“|”|：|《|》|u3000|（|）|-|・|①|‘|’|'|\u3000| |―|？|、|；",re.S)
    content=quote_pat.sub('',content)
    return content
```

```python
rebellion = get_corpus('rebellion.txt') # 作品《水浒传》
just_living = get_corpus('just_living.txt') # 作品《活着》
higanbana = get_corpus('higanbana.txt') # 作品《彼岸花》
remember = get_corpus('remember.txt') # 作品《追忆似水年华》
```

```python
import jieba
```

```python
s = '他唱起了旧日的歌谣，先是口依呀啦呀唱出长长的引子，接着出现两句歌词――皇帝招我做女婿'
seg_list = jieba.cut(s)
print("Full Mode: " + "/ ".join(seg_list))
```

```python
slp = snlp(s)
print(slp.words)
```

```python
s = '最好还能遭遇到喜欢的天气，喜欢的男人和女人。任何一件事情，只要心甘情愿，总是能够变得简单。不会有任何复杂的借口和理由。'
seg_list = jieba.cut(s)
print("Full Mode: " + "/ ".join(seg_list))
slp = snlp(s)
print(slp.words)
```

### 结论：

两种分词结果对比发现，没有采用自定义词典下，jieba分词比snownlp分词效果好一些

```python
s = '斯万长着鹰钩鼻，绿眼珠，脑门儿很高，头发黄得发红，剪成勃莱桑那样的发式,维吉尔（公元前70年―19年）：拉丁诗人。有关阿里斯泰的描述，见于他的诗作《农事诗》，也在特威克汉姆那里有出现过'
seg_list = jieba.cut(s)
print("Full Mode: " + "/ ".join(seg_list))
pseg_list = pseg.cut(s)
for w in pseg_list:
    print(w.word, "/", w.flag, ", ", end=' ')
```

```python

def to_train_txt(obj,filename):
    with open(data_path + 'temp.txt','w',encoding='utf-8') as f:
        f.write(obj)
    cnt = 0
    with open(data_path + filename,'w',encoding='utf-8') as q:
        print('prepare to save %s' % filename)
    with open(data_path + 'temp.txt','r',encoding='utf-8') as f:
        for line in f.readlines():
            if len(line) < 3:
                continue
            else:
                temp_seg = jieba.cut(line)
                temp = " ".join(temp_seg)
                with open(data_path + filename,'a',encoding='utf-8') as q:
                    q.write(temp)
            cnt += 1
            if cnt%1000 == 0:
                print('process %d lines' % cnt)
```
